# FOOD DELIVERY TIME

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("swiggy app.csv")
```

```python
print(df.columns)
```

```python
df.head()
```

```python
print(df["Delivery time"].unique())
```

## AVERAGE DELIVERY TIME BY CITY

```python
avg_delivery_city = df.groupby("City")["Delivery time"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,4))
sns.barplot(x=avg_delivery_city.index,
            y=avg_delivery_city.values)
plt.title("Average delivery time by city")
plt.ylabel("Minutes")
plt.xticks(rotation=45)
plt.show()           
```

## TOP 5 MOST ORDERED FOOD CATEGORIES


```python
top_food = df["Food type"].value_counts().head(5)
plt.figure(figsize=(6,4))
sns.barplot(x = top_food.values,
            y = top_food.index)
plt.title("Top 5 Ordered Food")
plt.xlabel("Order count")
plt.show()           
```

## RATING DISTRIBUTION 

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Avg ratings"],bins=10,kde=True)
plt.title("Ratings Distribution By Customer")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

## ORDERS PER CUSTOMER

```python
orders_per_customer = df["ID"].value_counts()
plt.figure(figsize=(6,4))
sns.histplot(orders_per_customer,bins=30)
plt.title("Order Per Customer")
plt.xlabel("Number of orders")
plt.ylabel("Customers")
plt.show()
```
